## 正太分布进行缺失值填充

```python
import numpy as np
a1= np.array([0,0,5,5])
a2= np.array([2,2,3,3])

print(a1.mean(), a2.mean()) # 2.5 2.5
print (a1.var() , a2.var()) # 6.25 0.25
print (a1.std() ,a2.std())  # 2.5 0.5
```

```python
# 需要定义一个函数，给定列名和缺失值种类，返回填充好的数据
import copy

col_name = 'age'
test_df = copy.deepcopy(raw_ext_new_feature_label_resume_type)

def impute_missing_value_and_normalization_fun(test_df, col_name):
    col_value_list = []

    # 计算列的均值和方差
    for i in range(len(test_df)):
        col_value = test_df.loc[i, col_name]
        if pd.isnull(col_value):
            continue
        col_value_list.append(col_value)
    
    col_value_list = np.array(col_value_list)
    mu = col_value_list.mean()
    sigma = col_value_list.std()
    print(mu, sigma)

    # 填充列的缺失值
    for i in range(len(test_df)):
        col_value = test_df.loc[i, col_name]
        if pd.isnull(col_value):
            test_df.loc[i, col_name] = np.random.normal(mu, sigma)
        
    # 对列进行正态分布归一化
    for i in range(len(test_df)):
        col_value = test_df.loc[i, col_name]
        test_df.loc[i, col_name] = (col_value - mu)/sigma

    return test_df, mu, sigma

test_df, mu, sigma = impute_missing_value_and_normalization_fun(test_df, col_name)
```

```python
# 分母为sigma*sigma，明显看出来返回的数据不是标准正态分布
(np.random.normal(mu, sigma, 15) - mu)/sigma*sigma
```
